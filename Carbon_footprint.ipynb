```python
# Drop last column
df_gen.drop(['Country Code', 'Indicator Name', 'Indicator Code','Unnamed: 65'], axis=1, inplace=True)
df_gen.shape

#transforming date columns to to one columns "year"

df_gen2 = pd.melt(df_gen, id_vars= ['Country Name'], 
                  var_name="Year", value_name="emmission value")


#filling empty values with zeros

df_gen2['emmission value'] = df_gen2['emmission value'].fillna(0)


#extracting the united states data only

ame_gen = df_gen2[df_gen2['Country Name'] == 'United States']
ame_gen.shape
```

```python
# Using graph_objects
import plotly.graph_objects as go

fig = go.Figure([go.Scatter(x=ame_gen['Year'], y=ame_gen['emmission value'])])
fig.show()
```
